# checking unique walues of each features

```python
unique_val = df['Sex'].unique()
print("unique values of :" ,unique_val)

#now encodeing it with lableEncoder

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

gender_encoded = le.fit_transform(df['Sex'])
df['gender_encoded'] = gender_encoded
df
```

```python
unique_val = df['ChestPainType'].unique()
print("unique values of ", unique_val)

#encodeing with lableincod
# from sklearn.preprocessing import LabelEncoder
# le = LabelEncoder()
# ChestPainType_encode = le.fit_transform(df['ChestPainType'])
# df['chest_pain_encoded'] = ChestPainType_encode
# df['chest_pain_encoded'].unique()


from sklearn.preprocessing import OneHotEncoder
encode = OneHotEncoder(sparse_output=False)

chest_pain_encoded = encode.fit_transform(df[['ChestPainType']])
chest_pain_df = pd.DataFrame(chest_pain_encoded , columns=encode.get_feature_names_out())
# Align index
chest_pain_df.index = df.index

df = pd.concat([df.drop('ChestPainType', axis=1), chest_pain_df], axis=1)
```

```python
df
```

```python
unique_val = df['RestingECG'].unique()
print("unique values of ", unique_val)

# from sklearn.preprocessing import LabelEncoder
# le = LabelEncoder()

# ecg_encoded = le.fit_transform(df['RestingECG'])

# df['ecg_encoded'] = ecg_encoded
# df['ecg_encoded'].unique()


from sklearn.preprocessing import OneHotEncoder
encode = OneHotEncoder(sparse_output=False)
ecg_encode = encode.fit_transform(df[['RestingECG']])

ecg_feature_df = pd.DataFrame(ecg_encode, columns=encode.get_feature_names_out())
ecg_feature_df.index = df.index
```

```python
unique_val = df['ST_Slope'].unique()
print("unique values of ", unique_val)

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

st_sloop_encoded = le.fit_transform(df['ST_Slope'])
df['st_sloop_encoded'] = st_sloop_encoded
df['st_sloop_encoded'].unique()
```

```python

unique_val = df['ExerciseAngina'].unique()
print("unique values of ", unique_val)

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

excersise_encoded = le.fit_transform(df['ExerciseAngina'])
df['excersise_encoded'] = excersise_encoded
df['excersise_encoded'].unique()
```
