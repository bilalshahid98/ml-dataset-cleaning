```python
import pandas as pd 
```

```python
df = pd.read_csv("Superstore.csv", encoding='latin-1')
```

```python
df = df.drop('Postal Code' , axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Product Name'].unique()
```

```python
# now incoding a object data type
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# feature which need le = Country , Category
df['Country'] = le.fit_transform(df['Country'])
df['Category'] = le.fit_transform(df['Category'])
```

```python
df['Category']
```

```python
df.info()

df['Ship Mode'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
encode = OneHotEncoder(sparse_output=False)

ship_mode_encode = encode.fit_transform(df[['Ship Mode']])

ship_mode_df = pd.DataFrame(ship_mode_encode, columns=encode.get_feature_names_out())

df = pd.concat([df, ship_mode_df], axis=1)

df = df.drop('Ship Mode', axis=1)
```

```python
# now encodeing Segment feature with onehot
segment_encode = encode.fit_transform(df[['Segment']])

sagment_encode_df = pd.DataFrame(segment_encode , columns=encode.get_feature_names_out())

df = pd.concat([df, sagment_encode_df] , axis=1)
```

```python
df = df.drop('Segment' , axis=1)
```

```python
# now doing same for City
city_encode = encode.fit_transform(df[['City']])

city_encode_df = pd.DataFrame(city_encode , columns=encode.get_feature_names_out())

df = pd.concat([df , city_encode_df] , axis=1)

df = df.drop('City',axis=1)
```

```python
#now doing for State

state_encode = encode.fit_transform(df[['State']])

state_encode_df = pd.DataFrame(state_encode , columns=encode.get_feature_names_out())

df = pd.concat([df , sagment_encode_df] , axis=1)

```

```python
df = df.drop('State', axis=1)
```

```python
# encodeing Region 

region_encode = encode.fit_transform(df[['Region']])

region_encode_df = pd.DataFrame(region_encode , columns=encode.get_feature_names_out())

df = pd.concat([df , region_encode_df] , axis=1)
```

```python
df = df.drop('Region', axis=1)
```

```python
df.head()
```

```python
df['Sub-Category'].unique()
```

```python
sub_cat_encode = encode.fit_transform(df[['Sub-Category']])

sub_cat_encode_df = pd.DataFrame(sub_cat_encode , columns=encode.get_feature_names_out())

df = pd.concat([df , sub_cat_encode_df] , axis=1)
df= df.drop('Sub-Category' , axis=1)
```
